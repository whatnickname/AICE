# <b> 2. 데이터 선택하기

## <B> (1) 컬럼명으로 데이터 선택

```python
# CSV 파일 읽기
file_path = 'datasets/winequality-white.csv'
wine_data = pd.read_csv(file_path, sep=",")

# 'quality' 열 선택
quality_data = wine_data['quality']

# 데이터 확인
quality_data.head()
```

```python
# 'quality'와 'pH' 열 선택
selected_columns = wine_data[['quality', 'pH']]
selected_columns.head()
```

## <B> (2) 범위 지정

```python
# 데이터프레임의 상위 10개 행 선택
top_10_rows = wine_data[:10]
top_10_rows
```

## <B> (3) 특정 행, 열 선택

```python
# 첫 번째 행의 두 번째 열 데이터
value = wine_data.iloc[0, 1]
print(value)
```

```python
# 첫 번째 행에서 'quality' 열의 값 선택
value = wine_data.loc[0, 'quality']
print(value)
```

## <B> (4) 데이터 조건으로 선택

```python
# 예: 와인의 품질이 7 이상인 데이터 선택
high_quality = wine_data[wine_data['quality'] >= 7]
high_quality.head()
```

```python
#예 :품질이 8 이상이고 pH가 3 이상인 데이터
high_quality_alcohol = wine_data[(wine_data['quality'] >= 8) & (wine_data['pH'] >= 3)]
high_quality_alcohol.head()
```
