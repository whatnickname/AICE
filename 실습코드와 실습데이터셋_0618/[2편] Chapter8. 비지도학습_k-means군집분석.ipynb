```python
#  
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#   
np.random.seed(42)  #    
data_1 = np.random.normal(loc=[2, 2], scale=0.5, size=(100, 2))  #   
data_2 = np.random.normal(loc=[8, 8], scale=0.5, size=(100, 2))  #   
data_3 = np.random.normal(loc=[5, 2], scale=0.5, size=(100, 2))  #   

#   
data = np.vstack([data_1, data_2, data_3])

#  
plt.figure(figsize=(7, 7))
plt.scatter(data[:, 0], data[:, 1], color='gray', alpha=0.6)
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
# K-Means    
kmeans = KMeans(n_clusters=3, random_state=42)  #   
kmeans.fit(data)  #  

#  
labels = kmeans.labels_  #      
centroids = kmeans.cluster_centers_  #   
```

```python
#   
centroids = kmeans.cluster_centers_ 
centroids
```

```python
#   
plt.figure(figsize=(7, 7))
for cluster in np.unique(labels):
    plt.scatter(
        data[labels == cluster, 0],  #    x 
        data[labels == cluster, 1],  #    y 
        label=f'Cluster {cluster + 1}',  #  
        alpha=0.6
    )
#  
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 
# WCSS(Within-Cluster Sum of Squares)  
wcss = []

#  K   K-Means  
for k in range(1, 11):  # K  1 10 
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)  #  
    wcss.append(kmeans.inertia_)  # WCSS  

#   
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))  # x 1~10  
plt.grid(True)
plt.show()
```
