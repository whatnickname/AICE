## <b>matplotlib.pyplot: 가장 기본적인 시각화 도구

```python
# 시각화 라이브러리 임포트
# 시각화 출력시 경고메시지 안나오는 코드
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# ✅ 한글 폰트 설정 (Windows 기준: 'Malgun Gothic')
plt.rcParams['font.family'] = 'Malgun Gothic'
# ✅ 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 데이터셋 임포트
df = pd.read_csv("datasets/heart.csv")
df.head()
```

```python
# 1. 나이별 평균 콜레스테롤 계산
age_chol = df.groupby('age')['chol'].mean().reset_index()
# 2. 선 그래프 그리기
plt.plot(age_chol['age'], age_chol['chol'])
plt.xlabel('나이')
plt.ylabel('평균 콜레스테롤 수치')
plt.title('나이별 평균 콜레스테롤 변화')
plt.grid(True)
plt.show()
# plt.plot(x, y) 는 선을 그리는 기본 구조입니다.
```

```python
plt.bar(df['sex'], df['chol'])  # x축: 성별(0,1), y축: 평균 콜레스테롤
plt.xlabel('성별')
plt.ylabel('콜레스테롤 수치')
plt.title('성별에 따른 콜레스테롤 비교')
plt.show()
```

```python
plt.scatter(df['age'], df['thalachh'])  # x축: 나이, y축: 최대 심박수
plt.xlabel('나이')
plt.ylabel('최대 심박수')
plt.title('나이와 심박수의 관계')
plt.show()
```

## <b>seaborn: 더 예쁜 그래프, 통계 기반 시각화

```python
# 시각화 라이브러리 임포트
# 시각화 출력시 경고메시지 안나오는 코드
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# 한글 폰트 설정 (Windows 기준: 'Malgun Gothic')
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
import seaborn as sns

sns.lineplot(data=df, x='age', y='chol')
plt.title('나이에 따른 콜레스테롤 변화')
plt.show()
```

```python
sns.barplot(data=df, x='sex', y='chol')  # 평균 자동 계산
plt.title('성별에 따른 평균 콜레스테롤')
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='thalachh')
plt.title('나이와 최대 심박수 산점도')
plt.show()
```

```python
sns.histplot(data=df, x='age', bins=10, kde=True)
plt.title('나이 분포')
plt.show()
```

```python
sns.boxplot(data=df, x='sex', y='thalachh')
plt.title('성별에 따른 심박수 분포 (이상치 포함)')
plt.show()
```
